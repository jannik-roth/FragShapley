```python
import pandas as pd
```

```python
# From:
# In silico Prediction of Chemical Ames Mutagenicity
# Congying Xu, Feixiong Cheng, Lei Chen, Zheng Du, Weihua Li, Guixia Liu, Philip W. Lee, and Yun Tang
# Journal of Chemical Information and Modeling 2012 52 (11), 2840-2847
# DOI: 10.1021/ci300400a 
```

```python
mut = pd.read_excel('mutagenicity.xls', header=1)
```

```python
# generate nonstereo_aromatic_smiles
from rdkit.Chem import MolFromSmiles, MolToSmiles

def get_canonical_SMILES(smiles):
    try:
        mol = MolFromSmiles(smiles)
    except:
        return False
    if mol is None:
        return False
    return MolToSmiles(mol, canonical=True)

mut['nonstereo_aromatic_smiles'] = mut.SMILES.apply(get_canonical_SMILES)
```

```python
mut.loc[mut.nonstereo_aromatic_smiles == False].shape
```

```python
mut = mut.loc[mut.nonstereo_aromatic_smiles != False]
```

```python
mut.shape
```

```python
mut = mut.drop_duplicates(subset='nonstereo_aromatic_smiles')
```

```python
mut.head()
```

```python
mut['Labels'] = mut['Labels'].rename({'mutagens': 1,
                                      'non-mutagens': 0})
```

```python
mut = mut.rename(columns={'Labels': 'mutagenicity'})
```

```python
mut = mut[['nonstereo_aromatic_smiles', 'mutagenicity']]
```

```python
mut.to_csv('classification/mutagenicity.csv', index=False)
```
