```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
min, max = -2.5, 2.5
norm = mpl.colors.Normalize(vmin=min, vmax=max)

for name_cm in ['bwr', 'cool', 'coolwarm', 'spring', 'summer', 'Wistia']:
    cmap = cm.get_cmap(name_cm)

    fig, axs = plt.subplots(1, 1, figsize=(6, 0.8))
    gradient = np.linspace(min, max, 256)
    gradient = np.vstack((gradient, gradient))
    axs.imshow(gradient, aspect='auto', cmap=cmap, norm=norm)
    axs.axis('off')
    fig.tight_layout()
    fig.savefig(f'cb_{name_cm}.svg', bbox_inches='tight')
```
