```python
from rdkit import Chem

def as_list(x):
    if x is None:
        return []
    if isinstance(x, str):
        return [x]
    return x


def compile_toxicophores(toxicophores):
    compiled = []

    for tox in toxicophores:
        patterns = []
        exclusions = []

        for s in as_list(tox["smarts"]):
            patt = Chem.MolFromSmarts(s)
            if patt is None:
                print("Invalid SMARTS:", tox["name"], s)
            else:
                patterns.append(patt)

        for s in as_list(tox.get("exclusion")):
            excl = Chem.MolFromSmarts(s)
            if excl is None:
                print("Invalid exclusion:", tox["name"], s)
            else:
                exclusions.append(excl)

        compiled.append({
            "name": tox["name"],
            "patterns": patterns,
            "exclusions": exclusions,
        })

    return compiled
```

```python
compiled_toxicophores = compile_toxicophores(toxicophores)
```

```python
def contains_toxicophore(smiles, compiled_toxicophores):
    mol = Chem.MolFromSmiles(smiles)

    if mol is None:
        return False

    for tox in compiled_toxicophores:

        has_pattern = any(
            mol.HasSubstructMatch(pattern)
            for pattern in tox["patterns"]
        )

        if not has_pattern:
            continue

        has_exclusion = any(
            mol.HasSubstructMatch(exclusion)
            for exclusion in tox["exclusions"]
        )

        if not has_exclusion:
            return True

    return False
```

```python
from sklearn.metrics import precision_score, recall_score, roc_auc_score, matthews_corrcoef
metrics = ['precision', 'recall', 'roc_auc', 'mcc']
metric_funcs = [precision_score, recall_score, roc_auc_score, matthews_corrcoef]
```

```python
# Now evaluate performance, easiest to load the explanations and then go through the seeds by hand

results = []

model = 'rf'

for model in ['rf', 'gcn']:

    df = pd.read_pickle(f'../4_mutagenicity/{model}_classification_hansen/df_explanation.pkl')
    df['contains_toxicophore'] = df['smiles'].apply(lambda s: contains_toxicophore(s, compiled_toxicophores))
    for split in df.split.unique():
        df_tmp = df[df.split == split].copy()

        for metric, func in zip(metrics, metric_funcs):
            
            y_true_tox = df_tmp.loc[df_tmp.contains_toxicophore].y_true
            y_pred_tox = df_tmp.loc[df_tmp.contains_toxicophore].y_pred
            scores_tox = func(y_true_tox, y_pred_tox)

            y_true_no_tox = df_tmp.loc[~df_tmp.contains_toxicophore].y_true
            y_pred_no_tox = df_tmp.loc[~df_tmp.contains_toxicophore].y_pred
            scores_no_tox = func(y_true_no_tox, y_pred_no_tox)
            results.append({
                'model':model,
                'split': split,
                'metric': metric,
                'score_contains_toxicophore': scores_tox,
                'score_no_toxicophore': scores_no_tox,
            })
```

```python
out = pd.DataFrame(results)
```

```python
summary = (
    out.groupby(["model", "metric"])
      .agg(
          toxicophore_mean=("score_contains_toxicophore", "mean"),
          toxicophore_std=("score_contains_toxicophore", "std"),
          no_toxicophore_mean=("score_no_toxicophore", "mean"),
          no_toxicophore_std=("score_no_toxicophore", "std"),
      )
      .reset_index()
)
```
